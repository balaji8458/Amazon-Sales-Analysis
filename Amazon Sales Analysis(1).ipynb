## Monthly Order Trends 📆- Which help us for Planning Manpower and Resuouce

```python
data['Date'] = pd.to_datetime(data['Date'])  
data['Month'] =data['Month'] = data['Date'].dt.strftime('%B  %Y')
 # Example: 'January', 'February'

plt.figure(figsize=(12,6))
sns.lineplot(x=data['Month'].astype(str), y=data['Qty'], estimator='sum', linestyle='dotted',data=data, marker='o', color='red')

plt.xlabel('Month')
plt.grid(True,color='blue')
```

## Percentage of Cancellations by Category

```python
cancelled_counts = data[data['Status'] == 'Cancelled']['Category'].value_counts()
total_counts = data['Category'].value_counts()

cancelled_percentage = (cancelled_counts / total_counts * 100).fillna(0)

plt.figure(figsize=(12,6))
sns.barplot(x=cancelled_percentage.index, y=cancelled_percentage.values, palette='viridis')

plt.xlabel('Category')
plt.ylabel('Cancellation Percentage')
plt.title('Percentage of Cancelled Orders by Category')
plt.xticks(rotation=45)
plt.show()
```

```python

# Ensure Amount is numeric


# Calculate Total Sales Amount
data['Total_Sales'] = data['Qty'] * data['Amount']

# Group by Category to get total sales per category
category_sales = data.groupby('Category', as_index=False)['Total_Sales'].sum()

# Plot Sales Amount
plt.figure(figsize=(12,6))
sns.barplot(x='Category', y='Total_Sales', data=category_sales, palette='magma')

plt.xlabel('Category')
plt.ylabel('Sales Amount (₹)')  # Add currency if applicable
plt.title('Category-wise Total Sales Amount')
plt.xticks(rotation=45)

# **Fix Y-ticks to show full sales values**
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation

# Add Grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
